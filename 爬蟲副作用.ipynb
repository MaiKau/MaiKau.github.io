<h2>Layer1 to Layer2 URL</h2> 

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
chrome_driver_path = 'D:\ACT\chromedriver'
url = 'https://www.kingnet.com.tw/medicine/list?selectType=section'
driver = webdriver.Chrome()
driver.implicitly_wait(10)
driver.get(url)
results = []  #網頁元素
link2=[]      #存URL的陣列
pages = 384
for page in range(1, pages+1):
    print('Now is: Page ', page)
    print('Working...')
    #逐項抓成分
    results = driver.find_elements_by_xpath("//td/a[@href and @style='white-space:nowrap;']") 
    for result in results:
        link2.append(result.get_attribute('href'))
    if page == pages:
        break
    else:
        time.sleep(10)
        driver.find_element_by_xpath("//a[@id='page-next']").click()
    
print('===Final Results===')
links.extend(link2)
print(links)
```
